```python
import cv2
import os
import re
from PIL import Image
import PyPDF2
import fitz
import docx
from transformers import AutoModelForQuestionAnswering, AutoTokenizer, pipeline

def generate_answer(plain_text, ques_text):
    nlp = pipeline('question-answering', model='deepset/roberta-base-squad2', tokenizer='deepset/roberta-base-squad2')
    question_set = {
        'question':ques_text,
        'context':plain_text
    }
    results = nlp(question_set)
    return results['answer']

def extract_text_txt(uploaded_txt_file, downloaded_txt_file):
    with open(uploaded_txt_file) as intxt:
        data = intxt.read()

    data = re.findall('[aA-zZ]+', data)
    with open(downloaded_txt_file, 'w') as outtxt:
        outtxt.write('\n'.join(data))
    return ' '.join(data)

def extract_text_pdf(uploaded_pdf_file):
    with fitz.open(uploaded_pdf_file) as intxt:
        text = ""
        for page in intxt:
            text += page.getText()
    return text

def extract_text_docx(uploaded_docx_file):
    doc = docx.Document(uploaded_docx_file)
    fullText = []
    for para in doc.paragraphs:
        fullText.append(para.text)
    return ' '.join(fullText)

upload_path = ""
download_path = ""

format_type = input('Apply Q&A for? (Plain Text/Documents): ')

if format_type == "Plain Text":
    text = input("Enter your text here: ")
    ques_text = input("Enter your Question: ")
    if ques_text is not None and text is not None and ques_text != "" and text != "":
        ans = generate_answer(text, ques_text)
        print("Here's the answer")
        print(ans)
    elif (ques_text is None or ques_text == "") and (text is not None or text != ""):
        print('Please enter your question!')
    elif (ques_text is not None or ques_text != "") and (text is None or text == "" ):
        print('Please enter your plain text!')
    else:
        print('Text fields missing! ')

if format_type == "Documents":
    print('Supports all popular document formats - TXT, PDF, DOCX')
    document_path = input('Enter the file path: ')
    if os.path.exists(document_path):
        if document_path.endswith('.txt') or document_path.endswith('.TXT'):
            uploaded_txt_file = os.path.abspath(document_path)
            downloaded_txt_file = os.path.abspath(str("processed_"+os.path.basename(document_path)))
            txt = extract_text_txt(uploaded_txt_file,downloaded_txt_file)

        if document_path.endswith('.pdf') or document_path.endswith('.PDF'):
            uploaded_pdf_file = os.path.abspath(document_path)
            txt = extract_text_pdf(uploaded_pdf_file)

        if document_path.endswith('.docx') or document_path.endswith('.DOCX'):
            uploaded_docx_file = os.path.abspath(document_path)
            txt = extract_text_docx(uploaded_docx_file)

    else:
        print('Invalid file path!')

    ques_text = input("Enter your Question: ")
    if ques_text is not None and ques_text != "":
        ans = generate_answer(str(txt),ques_text)
        print("Here's the answer ")
        print(ans)
    else:
        print('Please enter your question!')
```
